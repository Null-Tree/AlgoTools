```python
import networkx as nx
import matplotlib.pyplot as plt

graph=nx.Graph()

nodes:list[str] = [
    "A",
    "B",
    "C",
    "D"
]

# [source, target]
edges:list[tuple[str]] = [
    ("B", "A"),
    ("D","B"),
    ("B","C"),
    ("C","D")
]

graph.add_nodes_from(nodes)

graph.add_edges_from(edges)

# makes shell layour for graph a
pos = nx.shell_layout(graph)
nx.draw_networkx_nodes(graph,pos)
nx.draw_networkx_edges(graph,pos)
nx.draw_networkx_labels(graph,pos)

# plt.show()
```

```python


d1=graph.nodes
print(d1)
# ['A', 'B', 'C', 'D']

d2=graph.edges
print(d2)
# [('A', 'B'), ('B', 'D'), ('B', 'C'), ('C', 'D')]

d3=graph.edges.data()
print(d3)
# [('A', 'B', {}), ('B', 'D', {}), ('B', 'C', {}), ('C', 'D', {})]

d4=graph.adjacency
print(d4)
# <bound method Graph.adjacency of <networkx.classes.graph.Graph object at 0x000002A304057620>>

d5=graph.get_edge_data
print(d5)
# <bound method Graph.get_edge_data of <networkx.classes.graph.Graph object at 0x000002A304057620>>
```
